```python
# import libraries
import numpy as np
import pandas as pd
#import dataset
col_names=['last_name','date_of_birth','gender','first_name']
data = pd.read_csv('desktop/Deduplication Problem - Sample Dataset.csv',names=col_names)
data.head()
```

```python
data=data.drop(0,axis=0)
data.head()
```

```python
# describe total number of rows and columns 
data.describe(include='all')
```

```python
#calculate duplicate row in data set
data.duplicated().sum()
```

```python
data['first_name'].duplicated().sum()
```

```python
data['last_name'].duplicated().sum()
```

```python
# first we will remove duplicated rows along with all columns
data.drop_duplicates(keep = 'first', inplace = True)
data.head()
```

```python
# now we have only 62 rows and 4 columns
data.describe()
```

```python
# create full name by adding first_name and last_name column
data['full_name']=data['first_name']+" "+data['last_name']
data.head()
```

```python
data.drop_duplicates(subset = ['full_name'], keep = 'first', inplace = True)
data.describe()
```

```python
# as from the data set almost each name contain JR which means a young or baby.JR is used  in the United States to
# distinguish a son from his father, when both bear the same name.but by analyse the dataset that JR containing full_name and 
# JR not containing full_name has same date of birth which is impossible because fathers and sons date_of_birth cannot be same 
# so JR is not useful .We can remove it from dataset for better interpretation.
new_fullname = []
for name in data['full_name']:
    nwords = str.split(name)
    for word in nwords:
        if word== "JR" :
            nwords.remove(word)
    new_fullname.append(" ".join(nwords))
data['new_fullname'] = new_fullname
```

```python
data.head()
```

```python
data.drop_duplicates(subset = ['new_fullname'], keep = 'first', inplace = True)
data=data.drop(data[['last_name','first_name','full_name']],axis=1)
```
